```python
import re

log_entry ='2026-07-03 14:23:11,342 - ERROR - payment.service - Transaction failed for user_id=48213, amount=129.99, reason="insufficient_funds"'

pattern = r"- (?P<level>\w+) - .*?user_id=(?P<user>\d+).*?amount=(?P<amount>[\d.]+)"

match = re.search(pattern, log_entry)

if match:
    print(f"Full match: {match.group(0)}")
    print(f"Level: {match.group('level')}")
    print(f"User: {match.group('user')}")
    print(f"amount: {match.group(3)}")
    print(f"Group dict: {match.groupdict()}")
```

```python
import re

data = "item1, item2; item3, item4; item5"

pattern1 = r"\s*[,:]\s*"

data.split(",")
print(f"Character class split: {re.split(pattern1, data)}")

# delimiter 캡쳐

pattern2 = r"\s*([,:])\s*"
print(f"Character class split: {re.split(pattern2, data)}")

html = """
<p class = 'hello'> First para</p>
<b class = 'world'> second para</p>
"""

pattern3 = r"<.*?class = '(?:hello|world)'.*>"
print(f"HTML non-capturing split: {re.split(pattern3, html)}")
```

```python
import csv
from pathlib import Path

records = [
    {
        "host":"web01",
        "port":"80",
        "status": "running"
        
    },
    {
        "host":"db02",
        "port" : "5432",
        "status": "maintenance",
        "tags" : "prod, finance"
        
    }
]
    
out_dict_path = Path("output_dict.csv")
fieldnames = set()
    
for record in records:
        fieldnames = fieldnames | record.keys()
    
with out_dict_path.open("w", encoding = "utf-8", newline="") as file:
        writer = csv.DictWriter(
            file, fieldnames=
            fieldnames,
            restval='undefined',
            extrasaction="ignore")
        writer.writeheader()
        writer.writerows(records)
```
