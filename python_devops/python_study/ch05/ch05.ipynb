```python
import subprocess
import sys 
from time import time

def run_sleep(period):
    # Windows/Linux 어디서나 파이썬 인터프리터를 실행하여 0.1초 대기하도록 명령
    proc = subprocess.Popen(
        [sys.executable, '-c', f'import time; time.sleep({period})']
    )
    return proc

start = time()
procs = []
for _ in range(10):
    proc = run_sleep(0.1)
    procs.append(proc)
    
for proc in procs:
    try:
        proc.communicate(timeout=0.001)
    except subprocess.TimeoutExpired:
        proc.terminate()
        proc.wait()
        print("timeout")
    
    
end = time()

print(f"모든 프로세스가 종료하기 에 걸린 시간 {end - start}")
print('Exit status', proc.poll())
```

```python
from threading import Thread
from time import time

def factorize(number):
    for i in range(1, number +1):
        if number % i == 0:
            yield i
            
numbers = [2139079, 14245,32424,5345345]

class FactorizeThread(Thread):
    def __init__(self,number):
        super().__init__()
        self.number = number
    
    def run(self):
        self.factors = list(factorize(self.number))
        
start = time()
threads = []
for number in numbers:
    thread = FactorizeThread(number)
    thread.start()
    threads.append(thread)
for thread in threads:
    thread.join()
end = time()
print('Took %.3f seconds ' % (end - start))
```

```python
from collections import deque
from threading import Thread, Lock

def download(item):
    value = "다운로드하는 기능"
    return value

def resize(item):
    value = "조절하는 기능"
    return value
def upload(item):
    value = "업로드하는 기능"
    return value

# 스레드 안전 큐 : 스레드 간의 데이터를 주고받기 위한 공유자원
class MyQueue(object):
    def __init__(self):
        self.items = deque()
        self.lock = Lock()
        
    def put(self,item):
        with self.lock:
            self.items.append(item)
            
    def get(self):
        with self.lock:
            return self.items.popleft()

# 메인 작업 worker   
class Worker(Thread):
    def __init__(self, func, in_queue, out_queue):
        super().__init__()
        self.func = func
        self.in_queue = in_queue
        self.out_queue = out_queue
        self.polled_count = 0
        self.work_done = 0
        
    def run(self):
        while True:
            self.polled_count +=1
            try:
                item = self.in_queue.get()
            ## 이전 단계에서 작업이 완료되지 않아 입력 큐가 비어있는 경우
            except IndexError:
                sleep(0.01)
            else:
                result = self.func(item)
                self.out_queue.put(result)
                self.work_done += 1
                
download_queue = MyQueue()
resize_queue = MyQueue()
upload_queue = MyQueue()
done_queue = MyQueue()
threads = [
    Worker(download, download_queue, resize_queue),
    Worker(resize, resize_queue, upload_queue),
    Worker(upload, upload_queue, done_queue)
]

for thread in threads:
    thread.start()
for _ in range(1000):
    download_queue.put(object())
    
while len(done_queue.items) < 1000:
    sleep(0.01)
    
    
processed = len(done_queue.items)
polled = sum(t.polled_count for t in threads)
print('Processed', processed, 'items after polling', polled, 'times')
    
```

```python
def download(item):
    value = "다운로드하는 기능"
    return value

def resize(item):
    value = "조절하는 기능"
    return value
def upload(item):
    value = "업로드하는 기능"
    return value

class ClosableQueue(Queue):
    
    SENTINEL = object()
    
    def close(self):
        self.put(self.SENTINEL)
        
    def __iter__(self):
        while True:
            item = self.get()
            try:
                if item is self.SENTINEL:
                    return # 스레드 종료
                yield item
            finally:
                self.task_done()
                
class StoppableWorker(Thread):
    def __init__(self,func,in_queue,out_queue):
        super().__init__()
        self.func = func
        self.in_queue = in_queue
        self.out_queue = out_queue
        self.polled_count = 0
        self.work_done = 0
        
    def run(self):
        for item in self.in_queue:
            result = self.func(item)
            self.out_queue.put(result)
            
            
               
download_queue = ClosableQueue()
resize_queue = ClosableQueue()
upload_queue = ClosableQueue()
done_queue = ClosableQueue()
threads = [
    StoppableWorker(download, download_queue, resize_queue),
    StoppableWorker(resize, resize_queue, upload_queue),
    StoppableWorker(upload, upload_queue, done_queue)
]

for thread in threads:
    thread.start()
for _ in range(1000):
    download_queue.put(object())
download_queue.close()
download_queue.join()
resize_queue.close()
resize_queue.join()
upload_queue.close()
upload_queue.join()
print(done_queue.qsize(), 'item finished')
            
```
