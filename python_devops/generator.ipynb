```python
def count_up_to(limit):
    print("Generator function started....")
    n = 1
    
    while n <= limit :
        print(f"Yielding {n}")
        yield n
        print(f"Resumed after Yielding {n}")
        n += 1
        
    print("Generator function finished")

count_gen = count_up_to(3)
print(f"Returned object : {count_gen} of type {type(count_gen)}")

print("First call to next outside of for loop.")
next(count_gen)

print("Remaining output from for loop")
for number in count_gen:
    print(number)
```

```python
def get_list_of_server():
    print("Regular function started.")
    servers = []
    
    for i in range(3):
        
        server_name = f"server-{i}"
        print(f"\t\Adding {server_name}")
        servers.append(server_name)
    print("Regular function finished.")
    return servers
    
    
servers = get_list_of_server()
print(f"Returned list : {servers}")
```

```python
def yield_server(count):
    print("Generater function started.")

    
    for i in range(count):
        server_name = f"server-{i}"
        print(f"\tYielding {server_name}")
        yield server_name
        
    print("Generater function finished.")
        
servers_gen = yield_server(3) 
    
for server in servers_gen:  
    print(f"Returned list : {server}")
```

```python
import time

def load_data_eagerly(count,delay=0.1):
    
    result = []
    for i in range(count):
        time.sleep(delay)
        result.append(f"Data time {i}.")
        
    return result

t0 = time.time()
data = load_data_eagerly(5)
t1 = time.time()
print(f"ReGULA : Returned {data} in {t1-t0:.2f}s")

def load_data_lazily(count,delay=0.1):
    

    for i in range(count):
        time.sleep(delay)
        yield f"Data item {i}"
        
data_gen = load_data_lazily(5)

while True:
    t0 = time.time()
    try:
        item = next(data_gen)
    except StopIteration:
        break
    t1 = time.time()
    print(f"Received data {item} in {t1-t0:.2f}s")
```

```python
#1. Ingest the log lines
#2. Filter log lines based on either level or message substring
#3. Extract and return only the message attribute of the logs

import sys
import json

def read_logs(filepath):
    with open(filepath, 'r') as file:
        for line in file:
            line= line.strip()
            if not line:
                continue
            yield json.loads(line)
        
def filter_logs(logs, level=None, message_substring=None):
    for log in logs:
        if level is not None and log.get("level","").lower() != level.lower():
            continue
        if (message_substring is not None and message_substring.lower() not in log.get("message"," ").lower()):
            continue
        
        yield log    
        
def extract_field(logs, field="message"):
    for log in logs:
        yield log.get(field, " ").strip()
        
def get_first_n(logs, n =10):
    count = 0
    
    for log in logs:
        if count >= n:
            break
        
        yield log
        count += 1
        
logs_gen = read_logs("mock_system_log.txt")
filter_gen = filter_logs(logs_gen, "INFO", "user")
extract_gen = extract_field(filter_gen, "message")


for log in get_first_n(extract_gen,25):
    print(log)
    
print("Generator object sizes (in bytes):",
      sys.getsizeof(logs_gen),
      sys.getsizeof(filter_gen),
      sys.getsizeof(extract_gen),
      )
```

```python
import time

def simple_task():
    time.sleep(0.2)
    print("Running a smiple task....")
    
def timing_decorater(original_function):
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = original_function(*args, **kwargs)
        duration = time.perf_counter() - start
        print(f"{original_function.__name__} took {duration:.3f}s")
        
        return result
    return wrapper

simple_task = timing_decorater(simple_task)
simple_task()

```

```python
@timing_decorater
def another_task():
    print("Running another task...")
    
another_task()
```

```python
import random

def retry(max_attempts=3):
    def decorater(func):
        def wrapper(*args, **kwargs):
            for attempt in range(1, max_attempts+1):
                try:
                    print(f"Attempt {attempt}/{max_attempts}")
                    return func(*args, **kwargs)
                except Exception as e:
                    print(f"Error: {e}")
                    if attempt == max_attempts:
                        raise
                
        return wrapper
    return decorater

@retry(2)
def sometimes_fails():
    if random.random() < 0.7:
        raise RuntimeError("Flaky failure")
    return "Sucess!"

print(f"Result:{sometimes_fails()}")
```

```python
def log_and_reraise(func):
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except Exception as err:
            print(f"[ERROR] {func.__name__} raised {err.__class__.__name__}")
            raise
    return wrapper

@log_and_reraise
def fail():
    raise ValueError("simulated problem")

fail()
```

```python
def broken_decorator(func):
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs)
    return wrapper

@broken_decorator
def add(a,b):
    return a + b

print("Introspectation without @wraps:")
print(f" __name__:  {add.__name__}")
print(f" __doc__:  {add.__doc__}")
```

```python
from functools import wraps

def correct_decorator(func):
    @wraps(func)
    def wrapper(*args, **kwargs):
        return func(*args, **kwargs)
    return wrapper

@correct_decorator
def mul(a,b):
    """두 수의 곱 반환"""
    return a * b

print("Introspectation without @wraps:")
print(f" __name__:  {mul.__name__}")
print(f" __doc__:  {mul.__doc__}")
```
